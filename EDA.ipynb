```python
import ast
```

```python
def convert(text, key='name', limit=3):
    L = []
    counter = 0
    for item in ast.literal_eval(text):
        if counter < limit:
            L.append(item.get(key))
            counter += 1
    return L
```

```python
credits['cast'] = credits['cast'].apply(convert)
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
credits['crew'] = credits['crew'].apply(fetch_director)
```

```python
credits.head()
```

```python
keywords['keywords'] = keywords['keywords'].apply(convert)
keywords.head()
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
def clean(x):
    try:
        return float(x)
    except:
        return np.nan
```

```python
movies['popularity'] = movies['popularity'].apply(clean).astype('float')
movies['vote_count'] = movies['vote_count'].apply(clean).astype('float')
movies['vote_average'] = movies['vote_average'].apply(clean).astype('float')
```
